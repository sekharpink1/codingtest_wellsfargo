```python
# Importing the necessary libraries
import pandas as pd
```

```python
# Reading the sample files in to a dataframe.
df1 = pd.read_csv("/Users/sameer/Downloads/fwdcodingtestpythondeveloperwellsfargoopportunit/Input/data_source_1/sample_data.1.csv")
df2 = pd.read_csv("/Users/sameer/Downloads/fwdcodingtestpythondeveloperwellsfargoopportunit/Input/data_source_1/sample_data.2.dat",sep='|')
df3 = pd.read_csv("/Users/sameer/Downloads/fwdcodingtestpythondeveloperwellsfargoopportunit/Input/data_source_2/sample_data.3.dat")
```
